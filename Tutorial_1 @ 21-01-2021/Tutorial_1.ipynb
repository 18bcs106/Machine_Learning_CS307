```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
mu,sigma=5,1
x = np.random.normal(mu,sigma,10)
np.mean(x)
```

```python
def get_liklihood(z):
    ans= 1/(np.sqrt(2*np.pi*sigma**2)) * np.exp(-0.5*((x-z)/sigma)**2)
    return ans
```

```python
y = 1/(np.sqrt(2*np.pi*sigma**2)) * np.exp(-0.5*((x-mu)/sigma)**2)
y
```

```python
sns.lineplot(x,y,color="red")
plt.savefig("Guassian.png")
```

```python
tmp=np.linspace(1,10,12)
tmp
```

```python
plt.figure(figsize = (20,10))
for i in range(1,12,1):
    y1=get_liklihood(tmp[i])
    plt.subplot(3,4,i)
    sns.lineplot(x,y1,color="red")
    plt.title('Mean = '+ str(i) + "Variance = " + str(sigma))
    plt.xlabel('X')
    plt.ylabel('Y')
plt.savefig("Plot.png")
```
